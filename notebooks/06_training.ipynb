```python
import joblib
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import load_model

# Load data
X_train = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/X_train_fe.pkl")
X_test = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/X_test_fe.pkl")
y_train = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/y_train_fe.pkl")
y_test = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/y_test_fe.pkl")

# Load model
dl_model = load_model("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/models/dl_model.h5")

print("Data and model loaded successfully")
```

```python
dl_model = load_model("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/models/dl_model.h5")

dl_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',   # or your original loss
    metrics=['accuracy']
)
```

```python
# Enable Eager Execution
import tensorflow as tf

tf.config.run_functions_eagerly(True)
```

```python

print(tf.__version__)
```

```python
history = dl_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=32,
    verbose=1
)

print("Model training completed")
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])

plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/accuracy.png")
plt.show()

print("Accuracy graph generated and saved")
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(["Train", "Validation"])

plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/loss.png")
plt.show()

print("Loss graph generated and saved")
```

```python
dl_model.save("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/models/dl_model_trained.h5")

print("Trained model saved successfully")
```
