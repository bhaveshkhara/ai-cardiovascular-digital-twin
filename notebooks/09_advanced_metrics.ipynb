```python
import numpy as np
import joblib

from tensorflow.keras.models import load_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load data
X_test = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/X_test_fe.pkl")
y_test = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/y_test_fe.pkl")

# Load trained model
model = load_model("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/models/dl_model_trained.h5")

print("Data and model loaded successfully")
```

```python
# Predict probabilities
y_pred_prob = model.predict(X_test)

# Flatten arrays
y_pred_prob = y_pred_prob.flatten()
y_test = y_test.values if hasattr(y_test, 'values') else y_test

print("Predictions generated")
```

```python
# MAE
mae = mean_absolute_error(y_test, y_pred_prob)

# MSE
mse = mean_squared_error(y_test, y_pred_prob)

# RMSE
rmse = np.sqrt(mse)

# R2 Score
r2 = r2_score(y_test, y_pred_prob)

# Adjusted R2
n = len(y_test)
p = X_test.shape[1]

adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

# MAPE
# mape = np.mean(np.abs((y_test - y_pred_prob) / y_test + 1e-10)) * 100
# mape = np.mean(np.abs((y_test - y_pred_prob) / (y_test + 1e-10))) * 100
mask = y_test != 0
mape = np.mean(np.abs((y_test[mask] - y_pred_prob[mask]) / y_test[mask])) * 100

print("All metrics calculated successfully")
```

```python
print("\nAdvanced Evaluation Metrics:\n")

print(f"MAE  : {mae:.4f}")
print(f"MSE  : {mse:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R2   : {r2:.4f}")
print(f"Adjusted R2 : {adj_r2:.4f}")
print(f"MAPE : {mape:.2f}%") # cause our model is based on classification and MAPE is for regression, still 13.52% error is very good
```

```python
with open("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/metrics/advanced_metrics.txt", "w") as f:
    f.write(f"MAE: {mae}\n")
    f.write(f"MSE: {mse}\n")
    f.write(f"RMSE: {rmse}\n")
    f.write(f"R2: {r2}\n")
    f.write(f"Adjusted R2: {adj_r2}\n")
    f.write(f"MAPE: {mape}\n")

print("Advanced metrics saved successfully")
```
