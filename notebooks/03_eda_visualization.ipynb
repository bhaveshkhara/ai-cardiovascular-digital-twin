```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_name = r"C:\Users\Lenovo\Desktop\cardio_digital_twin_project\data\raw\heart.csv"
# Load dataset
df = pd.read_csv(file_name)

df.head()
```

```python
sns.countplot(x='target', data=df)
plt.title("Heart Disease Distribution")
plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/Heart_Disease_Distribution.png")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/Feature_Correlation_Heatmap.png")
plt.show()
```

```python
sns.histplot(data=df, x="age", hue="target", bins=30)
plt.title("Age Distribution vs Heart Disease")
plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/Age_Distribution_vs_Heart_Disease.png")
plt.show()
```

```python
sns.countplot(x='cp', hue='target', data=df)
plt.title("Chest Pain Type vs Heart Disease")
plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/Chest_Pain_Type_vs_Heart_Disease.png")
plt.show()
```

```python
sns.scatterplot(x='thalach', y='age', hue='target', data=df)
plt.title("Max Heart Rate vs Age")
plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/Max_Heart_Rate_vs_Age.png")
plt.show()
```

```python
sns.boxplot(x='target', y='chol', data=df)
plt.title("Cholesterol vs Heart Disease")
plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/Cholesterol_vs_heart_Disease.png")
plt.show()
```
