```python
import numpy as np
import joblib
from tensorflow.keras.models import load_model

# Load model and scaler
model = load_model("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/models/dl_model_trained.h5")
scaler = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/scaler_fe.pkl")

print("Model and scaler loaded successfully")
```

```python
# Example patient input

sample_input = np.array([[ 
    52, 1, 0, 125, 212, 0, 1, 168, 0, 1.0, 2, 2, 3,  # original features
    2,   # age_group
    0.0, # heart_stress
    26500 # bp_chol
]])

print("Sample input created")
```

```python
sample_scaled = scaler.transform(sample_input)

print("Input scaled successfully")
```

```python
prediction = model.predict(sample_scaled)

result = (prediction > 0.5).astype(int)

print("Prediction completed")
```

```python
if result[0][0] == 1:
    print("High Risk of Heart Disease")
else:
    print("Low Risk of Heart Disease")

print("Prediction Probability:", prediction[0][0])
```
