```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load data
X_test = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/X_test_fe.pkl")
y_test = joblib.load("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/data/processed/y_test_fe.pkl")

# Load trained model
model = load_model("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/models/dl_model_trained.h5")

print("Data and trained model loaded successfully")
```

```python
# Predict probabilities
y_pred_prob = model.predict(X_test)

# Convert to binary (0/1)
y_pred = (y_pred_prob > 0.5).astype(int)

print("Predictions generated")
```

```python
acc = accuracy_score(y_test, y_pred)

print("Model Accuracy:", acc)
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d')

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.savefig("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/plots/confusion_matrix.png")
plt.show()

print("Confusion matrix generated and saved")
```

```python
report = classification_report(y_test, y_pred)

print("Classification Report:\n")
print(report)
```

```python
with open("C:/Users/Lenovo/Desktop/cardio_digital_twin_project/outputs/metrics/classification_report.txt", "w") as f:
    f.write(report)

print("Classification report saved successfully")
```
